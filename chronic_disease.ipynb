### We have noted that these three columns are numerical data.So,We have to convert this Object Data type to Numerical Data Type

```python
def convert_text2num(data,column):
    data[column]=pd.to_numeric(data[column],errors='coerce')
for col in text_col:
    convert_text2num(data,col)
    print(f"{col} : {data[col].dtype}")
```

### Now We have to take the count of the null values in the Data Set

```python
misssing=data.isnull().sum()
misssing[misssing>0].sort_values(ascending=False).head(20)
```

### Replacing that Null values with the mean values or mode values based on the perception

```python
def mean_val_impute(data,column):
    mean_val=data[column].mean()
    data[column].fillna(value=mean_val,inplace=True)

def mode_val_impute(data, column):
    mode_vals = data[column].mode()
    if not mode_vals.empty:
        mode_val = mode_vals[0]
        data[column] = data[column].fillna(mode_val)
    else:
        print(f"Column {column} has no mode. Cannot impute.")
```

```python
data.columns
```

### If the column is not a object type.Then Replace the null values with the Mean values.

```python
num_col=[col for col in data.columns if data[col].dtype!='object']
for col_name in num_col:
   mean_val_impute(data,col_name)
```

### If the column is a object type.Then Replace the null values with the Mode values.

```python
cat_col=[col for col in data.columns if data[col].dtype=='object']
for cat_name in cat_col:
    mode_val_impute(data,cat_name)
```

## Importing the Model Selection From Scikit-Learn available in Python.
### Now the distribute the data as Training Data and Testing Data.

```python
from sklearn.model_selection import train_test_split
data['Class'] = data['Class'].astype(int)
y=data['Class']
x=data.drop("Class",axis=1)


```

### The dataset consists of 400 entries, which are split into training and testing sets. Specifically, 75% of the data (300 entries) is used for training purposes, while the remaining 25% (100 entries) is reserved for testing.

### Also Fit the Decision Tree Algorithm For the Training Data

```python
from sklearn.tree import DecisionTreeClassifier
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
print(f"'X' shape: {x_train.shape}")
print(f"'Xtest' shape: {x_test.shape}")
dct=DecisionTreeClassifier()
dct.fit(x_train,y_train)
```

### Now Predicting the result.

```python
y_pred_dct=dct.predict(x_test)
y_pred_dct
```

### After that Importing Classification Algorithms such as Naive Bayes,KNeighbors Classifier,Random Forest Algorithm,Support Vector Machines(SVC) etc,...

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import  KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

### To determine the best classification algorithm for this model, an empty list is created and populated with various classification algorithms, which are then evaluated to identify the one yielding the highest accuracy

```python
models=[]
models.append(('Naive Bayes',GaussianNB()))
models.append(('KNN',KNeighborsClassifier(n_neighbors=8)))
models.append(('RandomForestClassifier',RandomForestClassifier()))
models.append(('DecisionTreeClassifier',DecisionTreeClassifier()))
models.append(('SVM',SVC(kernel='linear')))
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix
```

```python
for name,model in models:
    print(name,model)
    print()
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    print("Confusion Matrix: ", confusion_matrix(y_test,y_pred))
    print("\n")
    print("accuracy_score: ", accuracy_score(y_test,y_pred))
    print("\n")
    print("precision_score: ", precision_score(y_test,y_pred))
    print("\n")
    print("Recall Score: ", recall_score(y_test,y_pred))
    print("\n")
    print("f1_score: ",f1_score(y_test,y_pred))
    print("\n")
```

# Conclusion

## Based on the comparative analysis, it is evident that the RandomForestClassifier algorithm yields the highest accuracy among the evaluated models. Therefore, RandomForestClassifier can be confidently selected as the optimal classification algorithm for this predictive model.
